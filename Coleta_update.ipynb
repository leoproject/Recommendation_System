```python
# !pip3 install youtube_dl
# !pip3 install pandas
import numpy as np
import youtube_dl
import pandas as pd
```

```python
ydl = youtube_dl.YoutubeDL()
```

```python
queries = ["machine+learning","data+science","kaggle"]
```

```python
results = []
for query in queries:
    r = ydl.extract_info("ytsearchdate10:{}".format(query), download=False)
    for entry in r["entries"]:
        entry['query'] = query 
    results += r['entries']
```

```python
len(results)
```

```python
df =pd.DataFrame(results)


# colunas = ['title','upload_date','view_count']
df['upload_date'] = pd.to_datetime(df["upload_date"])
df['tempo_desde_pub'] = (pd.to_datetime("2022-12-03") - df['upload_date'])/ np.timedelta64(1,'D')
```
